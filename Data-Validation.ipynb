```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
data = pd.read_csv("Data/diabetes_012_health_indicators_BRFSS2015.csv/diabetes_012_health_indicators_BRFSS2015.csv")
data
```

```python
data.isna().sum()
```

```python
data=data.astype("int")
data
```

```python
#Detecting Outliers for BMI
IQR = np.quantile(data["BMI"],0.75)-np.quantile(data["BMI"],0.25)
mean = np.mean(data["BMI"])
std = np.std(data["BMI"])
upperwhiskers = round(mean + (1.5*IQR),2)
lowerwhiskers = round(mean - (1.5*IQR),2)
BMI_outliers_df = data[(data["BMI"]>upperwhiskers)|(data["BMI"]<lowerwhiskers)]
sns.boxplot(data["BMI"])
plt.title("BMI-Outliers")
plt.savefig("Visuals/BMI_outliers_box.png")
plt.show()
sns.kdeplot(BMI_outliers_df["BMI"],fill="True")
plt.xlabel("BMI-Values")
plt.savefig("Visuals/BMI_outliers_kde.png")
plt.show()
#Removing BMI>65
data = data[data["BMI"]<65]
```

```python
data.to_csv("Data/datavalidation.csv",index=False)
```
